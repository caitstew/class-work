# 16-2: Sitka Death Valley Comparison

```python
import csv

from datetime import datetime

from matplotlib import pyplot as plt

def weather_info(filename, dates, highs, lows):
    with open(filename) as f:
        reader = csv.reader(f)
        header_row = next(reader)
    
        for row in reader:
            try:
                current_date = datetime.strptime(row[0], "%Y-%m-%d")
                high = int(row[1])
                low =int(row[3])
            except ValueError:
                print(current_date, "Missing Data")
            else:
                dates.append(current_date)
                highs.append(high)
                lows.append(low)

dates, highs, lows = [], [], []
weather_info("sitka_weather_2014.csv", dates, highs, lows)
fig= plt.figure(dpi=128, figsize=(10,6))
plt.plot(dates, highs, c="red", alpha=0.5)
plt.plot(dates, lows, c="blue", alpha=0.5)
plt.fill_between(dates, highs, lows, facecolor= "red", alpha=0.15) 
    
dates, highs, lows = [], [], []         
weather_info("death_valley_2014.csv", dates, highs, lows)
plt.plot(dates, highs, c="red", alpha=0.2)
plt.plot(dates, lows, c="blue", alpha=0.2)
plt.fill_between(dates, highs, lows, facecolor= "blue", alpha=0.05)


plt.title("Daily High and Low Temperatures, 2014", fontsize=24)
plt.xlabel(" ", fontsize=16)
plt.ylabel("Temperature (F)", fontsize=16)
plt.tick_params(axis="both", which="major", labelsize=16)

plt.show()
```

# 16-3: Rainfall
I tried to get a dataset from https://catalog.data.gov/dataset?tags=weather on Seattle, but it kept returning that all of the dates were missing and I wasn't sure how to format the dates with datetime to recognize them. So I just used Sitka and changed the data to precipitation. 

```python
import csv
from matplotlib import pyplot as plt
from datetime import datetime
filename = "sitka_weather_2014.csv"
with open(filename) as f:
    reader = csv.reader(f)
    header_row = next(reader)
    
    dates, rains = [], []
    for row in reader:
        try:
            current_date = datetime.strptime(row[0], "%Y-%m-%d")
            rain = float(row[19])
        except ValueError:
            print(current_date, "Missing Data")
        else:
            dates.append(current_date)
            rains.append(rain)

fig= plt.figure(dpi=128, figsize=(10,6))
plt.plot(dates, rains, c="blue", alpha=0.5)

plt.title("Sitka Rainfall", fontsize=24)
plt.xlabel(" ", fontsize=16)
plt.ylabel("Inches", fontsize=16)
plt.tick_params(axis="both", which="major", labelsize=16)
plt.show()
```

```python
import csv
from matplotlib import pyplot as plt
from datetime import datetime
filename = "seattle_rain.csv"
with open(filename) as f:
    reader = csv.reader(f)
    header_row = next(reader)
    
    dates, rains = [], []
    for row in reader:
        try:
            current_date = datetime.strptime(row[0], "%m/%d/%y")
            rain = float(row[1])
        except ValueError:
            print(current_date, "Missing Data")
        else:
            dates.append(current_date)
            rains.append(rain)

fig= plt.figure(dpi=128, figsize=(10,6))
plt.plot(dates, rains, c="blue", alpha=0.5)

plt.title("Observed Monthly Accumulation of Rain at Seattle Station 1", fontsize=16)
plt.xlabel(" ", fontsize=16)
plt.ylabel("Inches", fontsize=16)
plt.tick_params(axis="both", which="major", labelsize=16)
plt.show()
```

# 16-4 Explore

```python
import csv
from matplotlib import pyplot as plt
from datetime import datetime
filename = "seattle_rain.csv"
with open(filename) as f:
    reader = csv.reader(f)
    header_row = next(reader)
    
    dates, rains1, rains2, rains3, rains4 = [], [], [], [], []
    for row in reader:
        try:
            current_date = datetime.strptime(row[0], "%m/%d/%y")
            rain1 = float(row[1])
            rain2=float(row[2])
            rain3=float(row[3])
            rain4=float(row[4])
        except ValueError:
            print(current_date, "Missing Data")
        else:
            dates.append(current_date)
            rains1.append(rain1)
            rains2.append(rain2)
            rains3.append(rain3)
            rains4.append(rain4)

fig= plt.figure(dpi=128, figsize=(10,6))
plt.plot(dates, rains1, c="blue", alpha=0.2)
plt.plot(dates, rains2, c="blue", alpha=0.4)
plt.plot(dates, rains3, c="blue", alpha=0.6)
plt.plot(dates, rains4, c="blue", alpha=0.8)

plt.title("Observed Monthly Accumulation of Rain at Seattle Stations 1-4", fontsize=24)
plt.xlabel(" ", fontsize=16)
plt.ylabel("Inches", fontsize=16)
plt.tick_params(axis="both", which="major", labelsize=16)
plt.show()
```
