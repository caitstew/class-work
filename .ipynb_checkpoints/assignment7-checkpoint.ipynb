## 7-5: Handle Missing Data: fillna, dropna, interpolate

```python
import pandas as pd
df = pd.read_csv("7-5_weather_data.csv")
df
```

```python
import pandas as pd
df = pd.read_csv("7-5_weather_data.csv")
type(df.day[0])
```

```python
df = pd.read_csv("7-5_weather_data.csv", parse_dates=["day"])
type(df.day[0])
#note the change in output
```

```python
#make day index set
import pandas as pd
df = pd.read_csv("7-5_weather_data.csv", parse_dates=["day"])
df.set_index("day", inplace=True)
#inplace to modify original dataframe
df
```

```python
new_df = df.fillna(0) #for making a best guess on value
new_df
```

```python
#specifying fillna value by column
new_df = df.fillna({
    "temperature" : 0, 
    "windspeed" : 0, 
    "event" : "no event"
})
new_df
```

```python
#getting a better estimate, carrying forward value from previous day
new_df = df.fillna(method="ffill")
# ffill= forward fill
new_df
```

```python
new_df = df.fillna(method="bfill")
# bfill= backward fill
new_df
```

```python
new_df = df.fillna(method="bfill", axis="columns")
#This copies value from value to the right, goes horizontally
new_df
```

```python
new_df = df.fillna(method="ffill", limit=1)
# This stops a value from copying down more than once, wasn't happening in our set
new_df
```

```python
#getting an even better guess of missing value
#interpolate, fill missing values with mean of value directly before and after
new_df = df.interpolate()
new_df
#linear interpolation, default of interpolate is linear
```

```python
#now considering jump in missing days
#IMPORTANT FEATURE
new_df = df.interpolate(method="time")
new_df
```

```python
#dropping all rows with missing values
new_df = df.dropna()
new_df
```

```python
#dropping rows that have all three values missing
new_df = df.dropna(how = "all")
new_df
#note 1/9 is gone
```

```python
#"If I have at least 1 filled value, keep the row", alternative to how= "all"
new_df = df.dropna(thresh = 1)
new_df
```

```python
new_df = df.dropna(thresh = 2)
new_df
#now 1/6 is dropped too. Requires 2 valid values
```

```python
#how to add missing dates
dt = pd.date_range("01-01-2017", "01-11-2017")
#set date range
idx=pd.DatetimeIndex(dt)
#create a date tme index
df = df.reindex(idx)
#reindex with the new date time index
df
#then ypu'd use a fillna value to fill missing dates with an estimate
```

## Question One: 
### How many movies has each reviewer reviewed? What is the average number of reviews written by reviewers?

```python
import pandas as pd
df = pd.read_csv("reviews.csv", usecols=["critic", "publication"])
df.columns = ["critic", "number of reviews"]

df.pivot_table(index="critic", aggfunc="count")
print(df.pivot_table(index="critic", aggfunc="count").mean())
%matplotlib inline
df.pivot_table(index="critic", aggfunc="count").plot()
df.pivot_table(index="critic", aggfunc="count")
```

## Question Two:
### How many times did each publisher publish? Min? Max? Standard Deviation? Mean?
I wanted to do a question more dissimilar from the first one, but couldn't really think of something else. I considered looking at the dates, but they were incredibly messy in this dataset. I also initially wanted to see the occurences of fresh vs. rotten for each movie title. I could figure out how to group by movie but could not figure out how to look at occurences within a column applied to each grouping. I eventually had to move onto another idea. 

```python
import pandas as pd
df = pd.read_csv("reviews.csv", usecols=["publication", "title"])
df.set_index('publication')
df.columns = ["publication", "number of publications"]

df.pivot_table(index="publication", aggfunc="count")
print("Max:")
print(df.pivot_table(index="publication", aggfunc="count").max())
print("Min:")
print(df.pivot_table(index="publication", aggfunc="count").min())
print("Mean:")
print(df.pivot_table(index="publication", aggfunc="count").mean())
print("Standard Deviation:")
print(df.pivot_table(index="publication", aggfunc="count").std())
%matplotlib inline
df.pivot_table(index="publication", aggfunc="count").plot()
df.pivot_table(index="publication", aggfunc="count")
```
